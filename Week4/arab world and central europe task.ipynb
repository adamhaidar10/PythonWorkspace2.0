```python
df2 = pd.concat([df_arabworld, df_ceb], axis="columns")
```

```python
df2
```

```python
df2 = df2.drop("Country Name")
df2 = df2.rename(columns={7: "Arab World", 36: "Central Europe"})
```

```python
df2
```

```python
df2["Arab World"] = df2["Arab World"].astype(int)
df2["Central Europe"] = df2["Central Europe"].astype(int)
```

```python
df2
```

```python
pyplot.rcParams["figure.figsize"] = [16, 10]
pyplot.xticks(rotation=45, ha="right")
pyplot.xlabel("Years")
pyplot.ylabel("Hundreds of millions of subscribers")
pyplot.title("Mobile phone subscribers in Central Europe and the Baltics and the Arab World")
pyplot.plot(df2["Arab World"])
pyplot.plot(df2["Central Europe"])
pyplot.legend(["Arab World", "Central Europe and the Baltics"])
```
