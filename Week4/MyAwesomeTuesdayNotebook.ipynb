```python
import pandas as pd
```

```python
from matplotlib import pyplot
```

```python
# can plot sets of data
x = [1, 2, 3, 4, 5]
y = [10, 12, 8, 16, 23]
z = [12, 20, 15, 11, 9]
pyplot.plot(x, y)
pyplot.plot(x, z)
pyplot.legend(["y values", "z values"])
pyplot.xlabel("x values")
pyplot.ylabel("y and z values")
pyplot.title("My Little Graph")
pyplot.show()
```
